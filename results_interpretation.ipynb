# Интерпретация результата
Чтобы внедрить data-driven decisions, подготовим понятную интерпретацию результатов задачи.
```
Типичные пользователи сегментов для 3-х кластеров:
[[40. 15.  9.]
 [11. 30. 15.]
 [11. 99. 11.]]

Типичные пользователи сегментов для 4-х кластеров:
[[30. 10. 15.]
 [11. 99. 11.]
 [50. 20.  2.]
 [11. 30. 15.]]
```

Для трёх кластеров пользовательские сегменты угадываются легко. Остаётся скорректировать признаки на основе работы алгоритма:

- Базовый: среднее время сессии — 40 минут, средний чек — 15 тысяч рублей, 9 месяцев с момента регистрации на сервисе.
- Продвинутый: среднее время сессии — 11 минут, средний чек — 30 тысяч рублей, 15 месяцев с момента регистрации.
- Премиальный: среднее время сессии — 11 минут, средний чек — 99 тысяч рублей, 11 месяцев с момента регистрации.

В четырёх кластерах базовый сегмент с характеристиками [40. 15.  9.] разбился на две подгруппы: [50. 20.  2.] и [30. 10. 15.].

## Bеские аргументы за деление признаков на четыре кластера

#### 1) Время с момента регистрации пользователей для двух подгрупп разное
Действительно разное: 2 и 15 месяцев. Причём средний чек больше у новичков: есть смысл ввести тариф для начинающих.

#### 2) Четыре кластера видны на части проекций. В обученных на четырёх кластерах моделях центроиды находятся ровно в своих сегментах, на трёх — один центр лежит между кластерами.
Центроид между кластерами плохо описывает сегмент: пользователей с близкими значениями нет. В четырёх кластерах пользователи описаны хорошо.

#### 3) Среднее время сессии для двух подгрупп базового сегмента разное.
Оно больше у недавно зарегистрировавшихся на сервисе. Это логично: новички знакомятся с возможностями сервиса. Для них сто́ит ввести дополнительный сегмент.

**В библиотеке ```sklearn``` учтены все особенности работы алгоритма k-средних. Если он плохо подходит данным, например, с большой размерностью, можно изменить параметр ```max_iter```. Число итераций алгоритма выясняется вызовом атрибута ```n_iter_```.**
