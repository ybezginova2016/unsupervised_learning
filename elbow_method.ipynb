# Elbow method: Оптимальное число кластеров

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
```

```python
data = pd.read_csv('/Users/yuliabezginova/PycharmProjects/unsupervised_learning/segments.csv')
```

```python
K = range(1, 8)
for k in K:
    model = KMeans(n_clusters=k, random_state=12345)
    model.fit(data)
    print('Число кластеров:', k) 
    print('Значение целевой функции', model.inertia_)
```

### Для четырех кластеров значение целевой функции заметно меньше, чем для трех. Дальше разница между соседними значениями не такая явная. Посмотрим на визуализацию.

## 2 Оптимальное число кластеров

### 1) Выведите на экран значения целевой функции для количества кластеров — от 1 до 7. При обучении примените параметр random_state=12345.

```python
data = pd.read_csv('/Users/yuliabezginova/PycharmProjects/unsupervised_learning/segments.csv')

K = range(1, 8)
for k in K:
    model = KMeans(n_clusters=k, random_state=12345)
    model.fit(data)
    print('Число кластеров:', k) 
    print('Значение целевой функции', model.inertia_)
```

### Для четырех кластеров значение функции заметно меньше, чем для трех. Дальше разница между соседними значениями не такая явная. Посмотрим на визуализацию модели с четырьмя кластерами.

### 2) Обучите модель для четырёх кластеров. Центроиды укажите так: ['0 centroid', '1 centroid', '2 centroid', '3 centroid'].  Постройте диаграмму pairplot с полученными центроидами и заливкой для модели. При обучении примените параметр random_state=12345.

```python
data = pd.read_csv('/Users/yuliabezginova/PycharmProjects/unsupervised_learning/segments.csv')

# Обучение модели для 4-х кластеров
model = KMeans(n_clusters=4, random_state=12345)
model.fit(data)

centroids = pd.DataFrame(model.cluster_centers_, columns=data.columns)

data['label'] = model.labels_.astype(str)

centroids['label'] = ['0 centroid', '1 centroid', '2 centroid', '3 centroid']
data_all = pd.concat([data, centroids], ignore_index=True)

# Построение графика
sns.pairplot(data_all, hue='label', diag_kind='hist');
```

### Теперь везде порядок: все группы данных попали в кластеры.

### 3) Обучите модели для трёх и четырёх кластеров. Выведите на экран округлённые центроиды полученных моделей. При обучении примените параметр random_state=12345.

```python
data = pd.read_csv('/Users/yuliabezginova/PycharmProjects/unsupervised_learning/segments.csv')

# Обучение модели для 3-х кластеров
model = KMeans(n_clusters=3, random_state=12345)
model.fit(data)

print("Типичные пользователи сегментов для 3-х кластеров:")
print(np.round(model.cluster_centers_, 0))

# Обучение модели для 4-х кластеров
model = KMeans(n_clusters=4, random_state=12345)
model.fit(data)

print("Типичные пользователи сегментов для 4-х кластеров:")
print(np.round(model.cluster_centers_, 0))
```

### Все, как на графике: два кластера остались прежними, а третий - разделился на два. Причем значение среднего чека в новом кластере самое маленькое.
